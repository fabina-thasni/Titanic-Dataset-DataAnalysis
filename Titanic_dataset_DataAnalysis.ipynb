```python
import pandas as pd
```

```python
path="/content/Titanic-Dataset.csv"
```

```python
df=pd.read_csv(path)
```

```python
df
```

```python
#1.Get the details of passengers who survived and were younger than 18 years old.

df[(df['Survived']==1)&(df['Age']<18)]

```

```python
#2.Get the details of male passengers who were older than 40 years old.

df[(df["Sex"]=='male') & (df['Age']>40)]
```

```python
#3.Get the details of passengers who paid a fare greater than $30 and survived.

df[(df['Fare']>30) & (df['Survived']==1)]
```

```python
#4.Get the details of female passengers who did not survive.

df[(df['Sex']=='female') & (df['Survived']==0)]
```

```python
#5.Get the details of passengers who were in either the first or second class.

df[(df['Pclass']==1) | (df['Pclass']==2)]
```

```python
#6.Get the details of passengers who had more than two siblings/spouses aboard the Titanic and did not survive.

df[(df['SibSp']>2) & (df['Survived']==0)]
```

```python
#7.Get the details of passengers who were in the third class and were older than 30 years old.
df[(df['Pclass']==3) & (df['Age']>30)]
```

```python
#8.Get the details of passengers who were younger than 15 years old and had parents/children aboard.
df[(df['Age']<15) & (df["Parch"]>0)]
```

```python
#9.Get the details of male passengers who paid less than $10 for their fare.
df[(df['Sex']=="male") & (df['Fare']<10)]
```

```python
#10.Get the details of passengers who were in the third class, younger than 20, and had siblings/spouses aboard.

df[(df['Pclass']==3) & (df['Age']<20) & (df['SibSp']>0)]
```

```python
#11.Get the details of passengers who were either younger than 10 years or older than 60 years, and survived.

df[((df['Age']<10) | (df['Age']>60)) & (df['Survived']==1)]

```

```python
#12.Get the name,passenger id and Ticket number who were female and embarked in southampton.

df[['Name','PassengerId','Ticket']][(df['Sex']=='female') & (df['Embarked']=='S')]
```

```python
#13.Get the names of male passengers who were survived

survived_male=df['Name'][(df['Sex']=='male') & (df['Survived']==1)]
survived_male
```

```python
#14. Get the name, passengerId and cabin number who were in the 2nd class and the ticket fare greater than 50

details_sclass_fare=df[['Name','PassengerId','Cabin']][(df['Pclass']==2) & (df['Fare']>50)]
details_sclass_fare
```

```python
#15.Get the details of passengers who were embarked in Cherbourg or olderthan 50 years
Cherbourg_passengers=df[(df['Embarked']=='C') | (df['Age']>50)]
Cherbourg_passengers
```

```python
#16.Get the passenger details of age in between 3 to 10 years
child=df[(df['Age']>3) & (df['Age']<10)]
child
```

```python
#17.Get the details of  male passengers who were survived or female passengers who not survived
df[(df['Sex']=='male')&(df['Survived']==1) | (df['Sex']=='female')&(df['Survived']==0)]
```

```python
#18.Get the passengerId,name and class who were embarked in queenstown and survived

df[['PassengerId','Name','Pclass']][(df['Embarked']=='S') & (df['Survived']==1)]
```

```python
#19.Get the details of passengers who have siblings and spouses abroad is morethan 2 and parents and children abroad is lessthan 2.
parch_sibsp=df[(df['SibSp']>2) & (df['Parch']<2)]
parch_sibsp
```

```python
#20.Get the details of male passengers who were embarked in southampton
male_southampton_passengers=df[(df['Sex']=='male') & (df['Embarked']=='S')]
male_southampton_passengers
```

```python
#21.Get the name,age,ticket, cabin number and fare of female passengers who were embarked in queenstown
queenstown_female_passengers=df[['Name','Age','Ticket','Cabin','Fare']][(df['Sex']=='female')&(df['Embarked']=='Q')]
queenstown_female_passengers
```

```python
#22. Get the details of passengers in first class
fstclass_passengers=df[(df['Pclass']==1)]
fstclass_passengers
```

```python
#23.Get the details of female passengers who have morethan 3 siblings and spouses abroad
female_passengers=df[(df['Sex']=='female') & (df['SibSp']>3)]
female_passengers
```

```python
#24.Get the male passengers details who either youngerthan 18 and olderthan 60.
df[((df['Sex']=='male')&((df['Age']<18) | (df['Age']>60)))]
```

```python
#25.Get the details of passengers who were in 2nd class ,olderthan 50 and parents and child abroad.
scndclass_aged_passengers=df[((df['Pclass']==2) & (df['Age']>50)) & (df['Parch']>0)]
scndclass_aged_passengers
```
